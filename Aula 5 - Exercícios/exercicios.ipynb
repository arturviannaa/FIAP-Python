```python
points = int(input("Informe sua pontuação: "))
win_rate = float(input("Informe a taxa de vitória (%): "))
has_ban = input("Possui punições ativas? (sim/não): ").lower().strip()

MESSAGES = {
    "ban": "Rank bloqueado!",
    "bronze": "Rank bronze!",
    "prata": "Rank prata!",
    "ouro": "Rank ouro!"
}

if has_ban == "sim":
    rank = MESSAGES["ban"]

else:
    if points > 2500 and win_rate >= 75:
        rank = MESSAGES["ouro"]

    elif points > 1000 and win_rate >= 50:
        rank = MESSAGES["prata"]

    else:
        rank = MESSAGES["bronze"]

print(f"Veredito final: {rank}")
```
